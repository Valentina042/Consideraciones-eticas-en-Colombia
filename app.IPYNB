```python
import pandas as pd
import numpy as np

df['Género:'] = df['Género:'].map({'Masculino': 1, 'Femenino': 0})

# Assuming you have a DataFrame named 'df'
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()

print("Numeric Columns:")
print(numeric_columns)
```

```python
import seaborn as sns
specific_columns = ["¿En qué semestre(s) cursó esa(s) materia(s)?", "¿En qué semestre(s) cursó esa(s) materia(s)?2", "Género:"]
#sns.pairplot(data=df[specific_columns].head(), hue="Género:")
df[specific_columns].head()
```

```python
def count_chars(text, char):
    return text.count(char)

# Define the character you want to count
char_to_count = ';'
# Overwrite 'IntegerColumn' with its string representation
df['¿En qué semestre(s) cursó esa(s) materia(s)?'] = df['¿En qué semestre(s) cursó esa(s) materia(s)?'].astype(str)
df['¿En qué semestre(s) cursó esa(s) materia(s)?2'] = df['¿En qué semestre(s) cursó esa(s) materia(s)?2'].astype(str)

# Create a new column based on the count of the specified character
df['No. Materias de IA'] = df['¿En qué semestre(s) cursó esa(s) materia(s)?'].apply(lambda x: count_chars(x, char_to_count))
df['No. Materias ètica en IA'] = df['¿En qué semestre(s) cursó esa(s) materia(s)?2'].apply(lambda x: count_chars(x, char_to_count))

specific_columns = ["¿En qué semestre(s) cursó esa(s) materia(s)?", "¿En qué semestre(s) cursó esa(s) materia(s)?2",'No. Materias de IA','No. Materias ètica en IA', "Género:"]
df[specific_columns] 
```
